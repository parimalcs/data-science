```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 5, 3, 7, 6])
y2 = np.array([3, 4, 6, 2, 8])

print("X data:", x)
print("Y1 data:", y1)
print("Y2 data:", y2)
```

```python
# Line Graph
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='Series 1', color='blue', linestyle='-', marker='o')
plt.plot(x, y2, label='Series 2', color='red', linestyle='--', marker='x')

plt.title('Sample Line Graph')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bar Graph
plt.figure(figsize=(10, 6))
plt.bar(x, y1, label='Bar Series', color='green', edgecolor='black')

plt.title('Sample Bar Graph')
plt.xlabel('Category')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

First, let's create a sample 2D array using NumPy. We'll make it a 5x10 array to demonstrate 5 distinct subplots.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sample 2D array (5 rows, 10 columns)
data_2d = np.random.rand(5, 10) * 100 # Random data between 0 and 100
x_values = np.arange(data_2d.shape[1]) # X-axis values for each subplot

print("Generated 2D Array (first 3 rows):\n", data_2d[:3])
```

Now, let's create a figure with 5 subplots. We'll arrange them in a 3x2 grid, leaving one empty spot (or you could choose a 5x1 arrangement). Each subplot will display one row of the `data_2d` array.

```python
# Create a figure and a set of subplots (3 rows, 2 columns for 5 plots)
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 12))

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Plot each row of the 2D array on a separate subplot
for i in range(data_2d.shape[0]):
    ax = axes[i]
    ax.plot(x_values, data_2d[i, :], marker='o', linestyle='-', label=f'Row {i+1}')
    ax.set_title(f'Subplot {i+1}: Data from Row {i+1}')
    ax.set_xlabel('X-axis (Column Index)')
    ax.set_ylabel('Value')
    ax.legend()
    ax.grid(True)

# Hide any unused subplots (if nrows*ncols > number of plots)
for j in range(data_2d.shape[0], len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
